```python
import matplotlib.pylab as plt
import numpy as np
import random 

# Generate training data
N = 6
shift = 2
x = np.arange(N) + shift

random.seed(0)
y = []
for _ in range(N):
    y.append(random.uniform(30, 50))
y[0]=50.0

# Fit polynomial 
p = np.polyfit(x,y, 4)
print(p)

# prediction
def velocity(x,p):
    return p[0]*x**4 + p[1]*x**3 + p[2]*x**2 + p[3]*x  + p[4]
x_predict = np.linspace(2, 7)
y_predict = velocity(x_predict, p)

# Visualization
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.plot(x_predict, y_predict)

```

```python
import plotly.graph_objects as go
import numpy as np
np.random.seed(1)

N = 100
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
sz = np.random.rand(N) * 30

fig = go.Figure()
fig.add_trace(go.Scatter(
    x=x,
    y=y,
    mode="markers",
    marker=go.scatter.Marker(
        size=sz,
        color=colors,
        opacity=0.6,
        colorscale="Viridis"
    )
))

fig.show()
fig.write_image("fig1.svg")
```
